```python
import pickle
import gzip
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with gzip.open('new_data/svt_df.pickle','rb') as f:
    svt_df = pickle.load(f)
fan_df = pd.read_csv('new_data/fan_df.csv')
kbo_df = pd.read_csv('new_data/kbo_df.csv')
```

# 선수 분류

# 기준  
- 10승 이상
- 160? 이닝 이상
- 방어율 3.5 이하?
- war에 따라
- 재계약 여부 (보류)

```python
kbo_df.head()
```

```python
kbo_df['10w'] = kbo_df['W'].map(lambda x:1 if x>=10 else 0)
kbo_df['10w'].value_counts()
```

```python
kbo_df['160ip'] = kbo_df['IP'].map(lambda x:1 if x>=160 else 0)
kbo_df['160ip'].value_counts()
```

```python
kbo_df['3era'] = kbo_df['ERA'].map(lambda x:1 if x<= 3.5 else 0).value_counts()
kbo_df['3era'].value_counts()
```

# 간단한 시각화

```python
# 데이터프레임에서 상관 행렬 계산
correlation_matrix = kbo_df.corr()

# 히트맵 그리기
plt.figure(figsize=(12, 8))  # 히트맵 크기 설정
sns.heatmap(correlation_matrix,cmap='coolwarm', fmt=".2f")
plt.title("KBO Data Correlation Heatmap")
plt.show()
```
