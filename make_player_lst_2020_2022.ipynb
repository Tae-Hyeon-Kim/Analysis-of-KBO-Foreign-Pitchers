```python
import pandas as pd
```

```python
data = pd.read_csv('data/player_lst.txt', sep = "\t")
```

```python
data.columns = ['c1']
data = data[data['c1'].str.contains('-')]
data
```

```python
data['c1'].iloc[0].split('(')[-1][:-1]
```

```python
def kor(x):
    k_name = x.split('-')[0].strip()
    return k_name
def eng(x):
    name = x.split('-')[1]
    return name.split('(')[0].split(',')[0].lower().strip()
def team(x):
    return x.split('(')[-1][:-1].split(',')[0].strip()
def release_status(x):
    return 1 if '방출' in x else 0
    
data['kor_name'] = data['c1'].apply(kor)
data['eng_name'] = data['c1'].apply(eng)
data['team'] = data['c1'].apply(team)
data['release'] = data['c1'].apply(release_status)
data
```

```python
data = data[data['kor_name']!='']
data = data.drop('c1',axis=1)
data.reset_index(inplace=True,drop=True)
data
```

```python
for i in range(len(data)):
    if i <= 9:
        data.loc[i,'end_dt'] = '2019-12-31'
    elif i <= 19:
        data.loc[i,'end_dt'] = '2020-12-31'
    else:
        data.loc[i,'end_dt'] = '2021-12-31'
data
```

```python
data.to_csv('players_list.csv',index=False)
```
