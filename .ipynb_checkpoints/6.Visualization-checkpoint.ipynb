```python
basic_f = info_df[(info_df['type']=='fan')&(info_df['basic']==1)].index
```

```python
basic_f_df = fan_df[list(basic_f)+['class']]
print(basic_f_df.columns)
basic_f_df.head()
```

- 평균값으로 합치거나, 총 합으로 합치거나, 최빈값으로 합치거나 셋 중 하나

```python
def get_birth(x):
    season,age = basic_f_df.loc[basic_f_df['Name']==x,'Season'],basic_f_df.loc[basic_f_df['Name']==x,'Age']
    birth = int(season.iloc[0]) - int(age.iloc[0]) - 1
    return birth

def get_season_count(x):
    return basic_f_df['Name'].value_counts().loc[x]
    
basic_f_df.loc[:,'birth'] = basic_f_df['Name'].apply(get_birth)
basic_f_df.loc[:,'season_cnt'] = basic_f_df['Name'].apply(get_season_count)
basic_f_df.head()
# print(basic_f_df['Name'].apply(get_birth))
```

```python
basic_results = basic_f_df.groupby('Name').mean().reset_index()
print(basic_results.shape)
basic_results.head()
```

## cate1  
- 구종, 구속 관련

![](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FAR4IM%2FbtqD4UItdUX%2FiTm7H294viStKrk3NNUQ11%2Fimg.png)

```python
cate1_s = info_df[(info_df['type']=='svt')&(info_df['cate1']==1)].index
cate1_f = info_df[(info_df['type']=='fan')&(info_df['cate1']==1)].index
```

```python
cate1_s_df = svt_df[list(cate1_s)+['class']]
cate1_s_df.head()
```

```python
cate1_f_df = fan_df[list(cate1_f)+['class']]
cate1_f_df.head()
```

```python
cate1_s_df.groupby(['pitch_type','pitch_name'])['player_name'].count().sort_values(ascending=False)
```

```python
cate1_result = pd.DataFrame()
for name in cate1_s_df['player_name'].unique():
    tmp_sr = cate1_s_df[cate1_s_df['player_name']==name]['pitch_type'].unique()
    tmp_pitchs = list(set(tmp_sr) - set(['PO','IN','AB']))
    #print(name,' : ',len(tmp_pitchs))
    cate1_result = cate1_result.append({'name': name, 'pitch_cnt': len(tmp_pitchs)}, ignore_index=True)
    
cate1_result
```

```python
# 선수 별 구종 평균 구속을 구했을 때, 결측치가 적은 상위 5개 구종 : 패스트볼, 체인지업, 슬라이더, 싱커, 커브
speed_avg_df = cate1_s_df.groupby(['player_name','pitch_type'])['release_speed'].mean().unstack().reset_index()
speed_avg_df.columns = ['name', 'AB', 'CH', 'CS', 'CU', 'EP', 'FA', 'FC', 'FF', 'FO',
       'FS', 'IN', 'KC', 'PO', 'SI', 'SL']
speed_avg_df.isna().sum().sort_values()[:6]
```

```python
speed_avg_df[['name','FF','CH','SI','SL','CU']].head()
```

```python
# 실행 주의 (1번만 실행)
cate1_result = pd.merge(cate1_result,speed_avg_df[['name','FF','CH','SI','SL','CU']],on='name',how='outer')
print(cate1_result.shape)
cate1_result.head()
```
