```python
f = open("new_data/fangraph_crawling2.txt", 'r',encoding='UTF8')
lines = f.readlines()
url_lst = []
name_check = []
tmp_name_lst = []

for i in lines:
    if i.startswith('https'):
        url_lst.append(i)
        continue   
    name = i.strip()
    if name and '---' not in name:
        tmp_name_lst.append(name)
    if len(tmp_name_lst)==10:
        name_check.append(tmp_name_lst)
        tmp_name_lst = []
name_check.append(tmp_name_lst)
print(url_lst[0])
```

```python
for i in name_check:
    print(len(i))
```

```python
df_dict = {}
for key,url in enumerate(url_lst):

    response = requests.get(url)

    if response.status_code == 200:

        soup = BeautifulSoup(response.text, 'html.parser')
        script_tag = soup.find('script', id='__NEXT_DATA__')
        script_content = script_tag.string
        data=json.loads(script_content)

        try:
            data = json.loads(script_content)
            result_df=pd.DataFrame()   

            length = len(data['props']['pageProps']['dehydratedState']['queries'][0]['state']['data']['data'])
            for i in range(length):
                tmp_dict = data['props']['pageProps']['dehydratedState']['queries'][0]['state']['data']['data'][i]
                tmp_col = data['props']['pageProps']['dehydratedState']['queries'][0]['state']['data']['data'][i].keys()
                df=pd.DataFrame(columns=tmp_col)
                df=df.append(tmp_dict, ignore_index=True)
                tmp_df = pd.DataFrame([tmp_dict], columns=tmp_col)
                result_df = pd.concat([result_df, tmp_df], ignore_index=True)
            df_dict[key] = result_df
        except json.JSONDecodeError:
            print("JSON 데이터를 파싱할 수 없습니다.")
     
    else:
        print('HTTP 요청이 실패하였습니다.')

```

```python
df_dict[0].head()
```

```python
for key in range(len(df_dict.keys())):
    result_df = df_dict[key]
    for i in range(len(result_df)):
        name=result_df['Name'].astype(str).values[i]
        cleaned_string = re.sub(r'<.*?>', '', name)
        extracted_text = cleaned_string.strip()
        result_df['Name'][i]=extracted_text
        
    for i in range(len(result_df)):
        team=result_df["Team"].astype(str).values[i]
        if '<' not in team:
            continue
        else:
            cleaned_text = team.split('>')[-2].split('<')[0]
            result_df['Team'][i]=cleaned_text
```
