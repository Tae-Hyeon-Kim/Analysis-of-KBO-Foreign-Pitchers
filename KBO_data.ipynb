# 동명이인 전처리 및 이름 통일화

```python
df1 = pd.read_csv('new_data/kbo_data.csv')
df1.head()
```

```python
df2 = pd.read_csv('new_data/players_df_ver2.csv')
df2.head()
```

```python
df2.loc[df2['kor_name']=='채드 벨','kor_name'] = '채드벨'
df2.loc[df2['kor_name']=='팻 딘','kor_name'] = '팻딘'
df2.loc[df2['kor_name']=='트래비스','kor_name'] = '트레비스'
df2.loc[df2['kor_name']=='샘슨','kor_name'] = '키버스 샘슨'
df2.loc[df2['kor_name']=='수아레즈','kor_name'] = '앤드류 수아레즈'
```

```python
name_lst1 = list(df1['이름'].unique())
name_lst2 = list(df2['kor_name'].unique())
name_lst1.append('아드리안 샘슨')
name_lst1.append('알버트 수아레즈')
name_dict = {}
for n1 in name_lst1:
    if n1 == '브룩스':
        name_dict['애런 브룩스'] = n1
        continue
    for n2 in name_lst2:
        if n1 in n2:
            if n1 in ['켈리','수아레즈','샘슨']:
                name_dict[n2] = n2
            else:
                name_dict[n2] = n1
            break
```

```python
print(len(name_lst2),len(name_dict.keys()))
print(set(name_lst2)-set(name_dict.keys()))
```

```python
test = []
for i,j in name_dict.items():
    if j in i:
        test.append('T')
    else:
        test.append('F')
print('count T: ',test.count('T'))
print('count F: ',test.count('F'))
```

```python
df2['kor_name'] = df2['kor_name'].map(name_dict)
df2['kor_name'].head()
```
